```python
#3. Используя вкладку "Данные об аудитории", посчитайте, чему будет равен retention первого дня у пользователей, пришедших в продукт 1 ноября 
```

```python
df = pd.read_csv("Данные об аудитории.csv")
df['date'] = pd.to_datetime(df['date'])
nov1_users = df[df['date'] == '2023-11-01']['user_id'].nunique()
returned_users = df[
    (df['date'] == '2023-11-02') &
    (df['user_id'].isin(df[df['date']=='2023-11-01']['user_id']))
]['user_id'].nunique()

retention = returned_users / nov1_users

print(f"Nov1 Retention: {retention:.1%}")
```

```python
total_users = df['user_id'].nunique()
users_with_views = df[df['view_adverts'] > 0]['user_id'].nunique()

conversion = users_with_views / total_users
print(f"Конверсия в просмотр объявления: {conversion:.1%}")
```

```python
#8. Во вкладке "Данные АБ-тестов" результаты трех несвязанных АБ тестов для ARPU (общая выручка/общее количество пользователей).
#Посмотрите на результаты тестов и интерпретируйте их. Напишите значения p-value, которые вы получили.
```

```python
import pandas as pd
from scipy import stats

df = pd.read_csv("Данные АБ тестов.csv")

experiments = df['experiment_num'].unique()
for exp in experiments:
    data = df[df['experiment_num'] == exp]
    control = data[data['experiment_group'] == 'control']['revenue']
    test = data[data['experiment_group'] == 'test']['revenue']
    
    t_stat, p_value = stats.ttest_ind(test, control, equal_var=False)
    
    print(f'Эксперимент {exp}')
    print(f'p-value = {p_value}')
    print()
```
